```python
import numpy as np
import pandas as pd

input_filename = 'Analytics_Sample_Data_%283%29_%282%29.xlsx'
df = pd.read_excel(input_filename,sheet_name = 0)
```

```python
print(df.columns.values)
print(df.info())
df.head(5)
```

```python
# remove null 
custom_str = 'Customer placed order datetime'
count_null = df[custom_str].isnull().sum() 
print(count_null)

ordered_str = 'Placed order with restaurant datetime'
count_null = df[ordered_str].isnull().sum()
print(count_null)

driver_str = 'Driver at restaurant datetime'
count_null = df[driver_str].isnull().sum()
print(count_null)

delivered_str = 'Delivered to consumer datetime'
count_null = df[delivered_str].isnull().sum() + (df[delivered_str] == df[custom_str]).sum()
print(count_null)

print('all custum order has been delivered')
```

```python
#  Write a query to calculate the average earnings per hour by day of week
from datetime import datetime as dt

# reference starting date: 2019-01-01
ref_day = '2016-01-01'

df['weekday'] =  ( pd.to_datetime(df['Customer placed order datetime'], format='%d %H:%M:%S') \
                    + (pd.to_datetime(ref_day) - pd.to_datetime('1900-01-01') ) 
                 ).dt.weekday

# 0 = Monday, 6=Sunday
result = df.groupby('weekday').agg({'Order total': np.mean})
result
```

```python
#  convert to datetime formate

# reference starting date: 2019-01-01
df['Customer placed order datetime'] = pd.to_datetime(df['Customer placed order datetime'], format='%d %H:%M:%S')\
             + (pd.to_datetime(ref_day) - pd.to_datetime('1900-01-01') )

df['Placed order with restaurant datetime'] = pd.to_datetime(df['Placed order with restaurant datetime'], format='%d %H:%M:%S')\
             + (pd.to_datetime(ref_day) - pd.to_datetime('1900-01-01') )

df['Driver at restaurant datetime'] = pd.to_datetime(df['Driver at restaurant datetime'], format='%d %H:%M:%S')\
             + (pd.to_datetime(ref_day) - pd.to_datetime('1900-01-01') )

df['Delivered to consumer datetime'] = pd.to_datetime(df['Delivered to consumer datetime'], format='%d %H:%M:%S')\
             + (pd.to_datetime(ref_day) - pd.to_datetime('1900-01-01') )

df.head(5)
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

```python
# plot order amount along time
df['day'] = df[custom_str].dt.day
df['hour'] = df[custom_str].dt.hour
df['profit'] = df['Order total'] - df['Amount of discount'] + df['Amount of tip']

group_day_hour = df.groupby(['day','hour']).agg({'day': 'count'}).rename(columns={'day':'amount'}).reset_index()

#figure
plt.plot(group_day_hour.index, group_day_hour.amount)

```

```python
# check orders amount, mean profit, mean tips, delivery gap 


#daily result
aggregation_criterion = {custom_str: 'count',
                         'profit': 'mean',
                         'Amount of tip': 'mean'}

new_names = {custom_str:'customer_orders_count',
             'profit': 'average_profit',
             'Amount of tip': 'average_tips'}
                        
group_daily = df.groupby('day').agg(aggregation_criterion).rename(columns=new_names).reset_index()




# view result
#group_daily
```

```python
#plot figures of average profit

plt.plot(group_daily.day, group_daily.customer_orders_count)
plt.xlabel('day')
plt.ylabel('order amount')
plt.title('order amount per day')
plt.show()

# profit = order total - discount + tips
plt.plot(group_daily.day, group_daily.average_profit)
plt.xlabel('day')
plt.ylabel('average profit')
plt.title('average profit (order total - discount + tips) per day')
plt.show()

plt.plot(group_daily.day, group_daily.average_tips)
plt.xlabel('day')
plt.ylabel('average tips')
plt.title('average tips per day')
plt.show()
```

```python
## plot order total distribution 
print('maximum of Order total:', np.max(df['Order total']))
print('mean of Order total:', np.mean(df['Order total']) )

df['Order total'].plot.hist(alpha = 0.5, bins= 20)
plt.title('order total distribution')

#plot order total < 600 distribution
plt.figure()
df_trunked = df[df['Order total']<=600]
df_trunked['Order total'].plot.hist(alpha= 0.5, bins = 50)
plt.title('order total <=600 distribution')
```

```python
# dilivery time gap for orders

#only for asap

df['delivery_gap']= df[delivered_str] - df[custom_str]

time_gap = df[df['Is ASAP']==True].delivery_gap.dt.seconds / 3600

#plot distributions
plt.figure()
time_gap.plot.hist(alpha = 0.5, bins=50)
plt.title('delviery gap in hour')
print('time gap, average (in minutes):', time_gap.mean() * 60)


plt.figure()
time_gap[time_gap <= 5].plot.hist(alpha = 0.5, bins=50)
plt.title('delivery gap in hour < 5hours ')
print('time gap <=5hours, average(in minutes):', time_gap[time_gap <= 5].mean() * 60)
```
